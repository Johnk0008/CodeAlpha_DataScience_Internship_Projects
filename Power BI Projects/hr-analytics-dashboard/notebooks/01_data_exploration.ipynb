```python
# Data Exploration Notebook
import sys
sys.path.append('../')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from src.data_processing import HRDataProcessor

# Initialize and load data
processor = HRDataProcessor()
df = processor.generate_sample_data(1000)

# Basic exploration
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nBasic Info:")
print(df.info())
print("\nDescriptive Statistics:")
print(df.describe())

# Visualizations
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
df['department'].value_counts().plot(kind='bar')
plt.title('Employee Distribution by Department')

plt.subplot(2, 3, 2)
df['position'].value_counts().plot(kind='bar')
plt.title('Employee Distribution by Position')

plt.subplot(2, 3, 3)
df['attrition'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Attrition Distribution')

plt.tight_layout()
plt.show()
```
